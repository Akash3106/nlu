# Cyberbullying Classification with NLU

Racism and Sexism is a problem of increasing size and occurence.        
Fortunately we can leverage the structure of natural language with the latest deep learning algorithms with NLU in just one line.


The Cyblerbullyinh classifier model uses universal sentence embeddings and is trained with the classifierdl algorithm provided by Spark NLP.

```
import nlu
news_pipe  = nlu.load('classify.cyberbullying')
news_pipe.predict(['All women have to wear pretty clothes', 'All black people are good at math'])
```
